```python
train = pd.read_csv("./train_metadata.csv")
train_extra = pd.read_csv("./BirdClef2024_additional.csv", low_memory=False)
```

```python
train["filename"] = train["filename"].apply(lambda x: f"../dataset/birdclef-2024/train_audio/{x}")
```

```python
train_extra["filename"] = train_extra["file"].apply(lambda x: f"/{x}.wav")
train_extra["filename"] = train_extra["primary_label"]+train_extra["filename"]
train_extra["filename"] = train_extra["filename"].apply(lambda x: f"../dataset/birdclef-2024/train_audio/{x}")
train_extra["secondary_labels"] = '[]'
```

```python
train = train[["filename", "primary_label", "secondary_labels", "latitude", "longitude"]]
train_extra = train_extra[["filename", "primary_label", "secondary_labels", "lat", "lng"]]
```

```python
train_extra = train_extra.rename({"lat": "latitude", "lng": "longitude"}, axis='columns')
```

```python
train_all = pd.concat([train, train_extra]).reset_index(drop=True)
```

```python
train_all = train_all.fillna(-1)
```

```python
train_all["keep"] = train_all.apply(lambda row: 0 < row['latitude'] < 30 and 65 < row['longitude'] < 85, axis=1).astype(int)
```

```python
train_all.filename.iloc[0]
```

```python
import librosa
from tqdm import tqdm

duration = []
shape = []
i=0
for f1 in tqdm(range(len(train_all.filename)), total=len(train_all.filename)):
    try:
        wav, sr = librosa.load(train_all.filename.iloc[f1][25:], sr=32000)
        duration.append(len(wav)/sr)
        shape.append(wav.shape[0])
    except:
        duration.append(-1)
        shape.append(-1)
    i+=1
#     if i==1000:
#         break
    
```

```python
import numpy as np
np.unique(shape).shape
```

```python
train_all['duration'] = duration 
train_all['shape'] = shape 
```

```python
import librosa
from tqdm import tqdm

dup_list = []
for f1 in tqdm(range(len(train_all.filename)), total=len(train_all.filename)):
    try:
        wav, sr = librosa.load(train_all.filename.iloc[f1][25:], sr=32000)
        shape_ = wav.shape[0]
#         print(len(train_all[train_all['shape']==shape]))
        if train_all[train_all['shape']==shape_].filename.shape[0]!=1:
            print(train_all[train_all['shape']==shape_].filename.shape, shape_)
            for f2 in train_all[train_all['shape']==shape_].filename:
                if f2[25:] == train_all.filename.iloc[f1][25:]: continue
                wav2, sr2 = librosa.load(f2[25:], sr=32000)
                if np.sum(wav-wav2)==0:
                    print(train_all.filename.iloc[f1], f2)
                    dup_list.append((train_all.filename.iloc[f1][25:], f2[25:]))
    except:
        dup_list.append((-1, -1))
#         shape.append(-1)
    
```
